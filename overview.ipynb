```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle('stock_returns.pkl')
```

```python
# benchmark dax values
dax_ret_5y_wo_corona = df[df.symbol == '^GDAXI'].ret_5y_wo_corona[0]
dax_ret_1y_wo_corona = df[df.symbol == '^GDAXI'].ret_1y_wo_corona[0]
dax_since_corona = df[df.symbol == '^GDAXI'].ret_since_corona[0]
```

```python
df_plot = df

fig, axs = plt.subplots(2, sharex=True, figsize=(20,20))
axs[0].bar(df_plot.name, df_plot.ret_1y_wo_corona, color='tab:blue', label='1y ret w/o Corona')
axs[0].bar(df_plot.name, df_plot.ret_since_corona, color='tab:orange', label='ret since Corona')
axs[0].hlines(dax_ret_1y_wo_corona, xmin=0, xmax=len(df), colors='tab:blue', linestyles='solid', label='dax 1y ret w/o Corona')
axs[0].hlines(dax_since_corona, xmin=0, xmax=len(df), colors='tab:orange', linestyles='solid', label='dax since Corona')
axs[0].legend()
axs[0].set_title("1y returns and losses since Corona")

axs[1].bar(df_plot.name, df_plot.ret_5y_wo_corona, color='tab:blue', label='5y ret w/o Corona')
axs[1].bar(df_plot.name, df_plot.ret_since_corona, color='tab:orange', label='ret since Corona')
axs[1].hlines(dax_ret_5y_wo_corona, xmin=0, xmax=len(df), colors='tab:blue', linestyles='solid', label='dax 5y ret w/o Corona')
axs[1].hlines(dax_since_corona, xmin=0, xmax=len(df), colors='tab:orange', linestyles='solid', label='dax since Corona')
# axs[1].xticks(rotation=90)
for tick in axs[1].get_xticklabels():
    tick.set_rotation(90)
axs[1].legend()
axs[1].set_title("5y returns and losses since Corona")

plt.show()
```

# 1-year & 5-year outperformance

```python
df_plot = df[(df.ret_5y_wo_corona > dax_ret_5y_wo_corona) & (df.ret_1y_wo_corona > dax_ret_1y_wo_corona)]

fig, axs = plt.subplots(2, sharex=True, figsize=(20,20))
axs[0].bar(df_plot.name, df_plot.ret_1y_wo_corona, color='tab:blue', label='1y ret w/o Corona')
axs[0].bar(df_plot.name, df_plot.ret_since_corona, color='tab:orange', label='ret since Corona')
axs[0].hlines(dax_ret_1y_wo_corona, xmin=0, xmax=len(df), colors='tab:blue', linestyles='solid', label='dax 1y ret w/o Corona')
axs[0].hlines(dax_since_corona, xmin=0, xmax=len(df), colors='tab:orange', linestyles='solid', label='dax since Corona')
axs[0].legend()
axs[0].set_title("1y returns and losses since Corona")

axs[1].bar(df_plot.name, df_plot.ret_5y_wo_corona, color='tab:blue', label='5y ret w/o Corona')
axs[1].bar(df_plot.name, df_plot.ret_since_corona, color='tab:orange', label='ret since Corona')
axs[1].hlines(dax_ret_5y_wo_corona, xmin=0, xmax=len(df), colors='tab:blue', linestyles='solid', label='dax 5y ret w/o Corona')
axs[1].hlines(dax_since_corona, xmin=0, xmax=len(df), colors='tab:orange', linestyles='solid', label='dax since Corona')
# axs[1].xticks(rotation=90)
for tick in axs[1].get_xticklabels():
    tick.set_rotation(90)
axs[1].legend()
axs[1].set_title("5y returns and losses since Corona")

plt.show()
```
